```python
from utils import load_vector
data = load_vector('ATTPT02010CTReqZsinogram.dat')
```

```python
sino = data[:,2].reshape(int(max(data[:,0]))+1, int(max(data[:,1]))+1)
```

```python
from matplotlib import pyplot as plt
from inverse_radon import iradon
angles = list(map(int, set(data[:,0])))
reconstruction_fbp = iradon(sino.T, angles)

imkwargs = dict(vmin=-0.2, vmax=0.2)
fig, (ax1) = plt.subplots(1, 1, figsize=(6, 6), sharex=True, sharey=True)
ax1.set_title("Reconstruction\nFiltered back projection")
ax1.imshow(reconstruction_fbp[::-1], cmap=plt.cm.Greys_r)
plt.show()
```
